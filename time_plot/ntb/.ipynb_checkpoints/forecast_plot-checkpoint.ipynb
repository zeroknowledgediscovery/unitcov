```python
import pandas as pd
import numpy as np
from datetime import timedelta
import pylab as plt
%matplotlib inline

import seaborn as sns
```

```python
df_ours = pd.read_csv('forecast_ours_plain_1.csv', index_col=0)
```

```python
df_otherTeams_1 = pd.read_csv('forecast_otherTeam_1-wk.csv', index_col=0, parse_dates=True)
df_otherTeams_2 = pd.read_csv('forecast_otherTeam_2-wk.csv', index_col=0, parse_dates=True)
dates = pd.date_range(start='2020-04-11', freq='7D', periods=len(df_ours))
dates_1 = dates + timedelta(days=7)
dates_2 = dates + timedelta(days=14)
```

```python
cols = [x for x in df_ours.columns if x.endswith('1')]
df_ours_1 = df_ours[cols]
df_ours_1.index = dates_1

cols = [x for x in df_ours.columns if x.endswith('2')]
df_ours_2 = df_ours[cols]
df_ours_2.index = dates_2
```

```python
df_combined_1 = df_otherTeams_1.join(df_ours_1.loc['2020-07-11':, :]).drop('JHU-CSSE-DECOM', axis=1).dropna()
df_combined_2 = df_otherTeams_2.join(df_ours_2.loc['2020-07-11':, :]).drop('JHU-CSSE-DECOM', axis=1).dropna()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
for col in ['Columbia-UNC-SurvCon', 'QJHong-Encounter', 'ET1']:
    if col == 'grd1':
        continue
    err = (df_combined_1[col] - df_combined_1['grd1']).values
    mse = (err ** 2).mean()
    sns.distplot(err, ax=ax, label=f'{col}\nmse={mse:.2e}')
ax.legend()
ax.set_title('One-Week-Ahead Prediction Error', fontsize=15)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
for col in ['Columbia-UNC-SurvCon', 'QJHong-Encounter', 'ET2']:
    if col == 'grd2':
        continue
    err = (df_combined_2[col] - df_combined_2['grd2']).values
    mse = (err ** 2).mean()
    sns.distplot(err, ax=ax, label=f'{col}\nmse={mse:.2e}')
ax.legend()
ax.set_title('Two-Week-Ahead Prediction Error', fontsize=15)
```

## 1-week-ahead

```python
ylim = np.array([np.nanmin(df_otherTeams_1.values), np.nanmax(df_otherTeams_1.values)])
margin = (ylim[1] - ylim[0]) * .05
ylim = [ylim[0] - margin, ylim[1] + margin]
figsize = (15, 7.5)
```

```python
fig, ax = plt.subplots(1, 1, figsize=figsize)
df_ours_1.loc['2020-07-01':, 'grd1'].plot(ax=ax, marker='v', linewidth=5)
df_otherTeams_1.plot(ax=ax, marker='o')
ax.set_ylim(ylim)
```

```python
fig, ax = plt.subplots(1, 1, figsize=figsize)
df_ours_1.loc['2020-07-01':].drop('grd1', axis=1).plot(ax=ax, marker='o')
df_ours_1.loc['2020-07-01':, 'grd1'].plot(ax=ax, marker='v', linewidth=5)
ax.set_ylim(ylim)
```

```python
fig, ax = plt.subplots(1, 1, figsize=figsize)
df_ours_1.loc['2020-07-01':, 'grd1'].plot(ax=ax, linewidth=5)
df_otherTeams_1[['Columbia-UNC-SurvCon', 'QJHong-Encounter']].plot(ax=ax, marker='v')
df_ours_1.loc['2020-07-01':][['ET1']].plot(ax=ax, marker='o')
ax.set_title('One-Week-Ahead Prediction Error', fontsize=15)
# ax.set_ylim(ylim)
```
