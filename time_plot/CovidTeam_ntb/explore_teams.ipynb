## Find teams that has inc case for counties

```python
import pandas as pd
from glob import glob
import os
```

```python
data_folder = '/home/yihuang/Documents/Data/covid19-forecast-hub/data-processed/'
```

```python
teams = []
for folder in glob(f'{data_folder}/*'):
    if not os.path.isdir(folder):
        continue
    csv_fnames = sorted(glob(folder + '/2020*csv'))
    if len(csv_fnames) > 0:
        
        team = folder.split('/')[-1]
        
        csv_fname = csv_fnames[-1]
        df = pd.read_csv(csv_fname, dtype={'location': str})
        df = df[df.target == '1 wk ahead inc case']
        locations = df.location.unique()
        if '01001' in locations:
            start = csv_fnames[0].split('/')[-1][:10]
            end = csv_fnames[-1].split('/')[-1][:10]
            print(f'{team}: {start} to {end}\n')
            teams.append(team)
```

```python
teams = [
    'CEID-Walk',
    'COVIDhub-baseline',
    'COVIDhub-ensemble',
    'CU-nochange',
    'CU-scenario_high',
    'CU-scenario_low',
    'CU-scenario_mid',
    'CU-select',
    'Google_Harvard-CPF',
    'IowaStateLW-STEM',
    'JHUAPL-Bucky',
    'JHU_IDD-CovidSP',
    'LNQ-ens1',
    'OliverWyman-Navigator',
    'PandemicCentral-USCounty',
    'UCLA-SuEIR',
    'USC-SI_kJalpha',
]
```

```python
for team in teams:
    team_folder = f'{data_folder}/{team}'
    csv_fnames = sorted(glob(f'{team_folder}/2020*csv'))
    
    print(f'\n{team}')
    dfs = []
    for csv_fname in csv_fnames:
        df = pd.read_csv(csv_fname, dtype={'location': str})
        df = df[(df.target == '1 wk ahead inc case') & (df.type == 'point')]
        # print(df.head())
        df = df[df.location.str.len() == 5]
        date = csv_fname.split('/')[-1][:10]
        print(f'\t{date}: {len(df)}')
        
        if len(df) > 0:
            target_end_date = df.target_end_date.values[0]
            df = df.drop(['target', 'type', 'quantile', 'forecast_date'], axis=1)\
                .rename(columns={'value': target_end_date})\
                .set_index('location')
            print(df.head())
            dfs.append(df)
            #'location', 'target_end_date'
```
