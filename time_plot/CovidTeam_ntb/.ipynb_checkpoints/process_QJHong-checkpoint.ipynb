```python
import pandas as pd
import numpy as np
from glob import glob
```

```python
team = 'QJHong-Encounter'
data_folder = f'../CovidTeam_data/{team}/'
csv_fnames = sorted(glob(f'{data_folder}/*{team}.csv'))
```

```python
for w in [1, 2, 3, 4, 5]:
    
    target = f'{w} wk ahead inc case'
    dfs = []
    for csv_fname in csv_fnames:
        df_raw = pd.read_csv(csv_fname).drop('location', axis=1)

        # get point estimate
        df_pt = df_raw[ (df_raw.target==target) & (df_raw.type == 'point') ]\
            .drop(['quantile', 'type'], axis=1)\
            .set_index('forecast_date')
        
        # get quantile estimate
        df_qn = df_raw[ (df_raw.target==target) & (df_raw.type == 'quantile') ].copy()
        df_qn = pd.pivot_table(df_qn, columns='quantile', values='value', index='forecast_date')
        cols = [f'{df_qn.columns.name}_{x}' for x in df_qn.columns]
        df_qn = df_qn.rename(columns={old: new for old, new in zip(df_qn.columns, cols)})
        df_qn.columns.name = None

        dfs.append(pd.concat([df_pt, df_qn], axis=1))
    
    df_date = pd.concat(dfs, axis=0)
    for col in cols + ['value']:
        df_date[col] = df_date[col]
        
    df_date.to_csv(f"../CovidTeam_results/combined_{team}_{target.replace(' ', '-')}.csv")
    print(df_date)
```
